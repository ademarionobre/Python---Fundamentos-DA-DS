Vari√°veis s√£o *espa√ßos alocados na mem√≥ria RAM* para *guardar valores **temporariamente***. Em Python, esses espa√ßos n√£o precisam ser tipados, ou seja, a vari√°vel pode ser alocada sem especificar o tipo de dado que ela aguardar√°. O interpretador Python √© capaz de determinar o tipo de dado da vari√°vel com base no seu valor, ou seja, as vari√°veis s√£o tipadas dinamicamente nessa linguagem.

Usamos o hash # para criar coment√°rios de uma linha. Lembrando que em qualquer linguagem de programa√ß√£o, os coment√°rios n√£o s√£o nem interpretados, nem compilados.

```python
# Exemplos de vari√°veis
x = 10
nome = 'aluno'
nota = 8.75
fez_inscricao = True
```

```python
# Para saber o **tipo** de dado que uma vari√°vel guarda, podemos imprimir seu tipo usando a fun√ß√£o type():
print(type(x))
print(type(nome))
print(type(nota))
print(type(fez_inscricao))
```

Acima temos a c√©lula de entrada com comandos para imprimir na tela, os tipos das quatro vari√°veis que criamos anteriormente. Veja que foram impressas quatro diferentes classes de tipos de dados. A vari√°vel "x" √© do tipo inteira (int). A vari√°vel "nome" √© do tipo string (str). A vari√°vel "nota" √© do tipo ponto flutuante (float). A vari√°vel "fez_inscricao" √© do tipo booleana (bool).

Em Python, tudo √© objeto! Por isso os tipos de dados aparecem com a palavra "class", que significa classe.

Vamos solicitar que o usu√°rio digite um nome, para que possamos fazer uma sauda√ß√£o personalizada. A fun√ß√£o ``input()`` faz a leitura de um valor digitado:

```python
# Comando input
nome = input("Digite um nome: ")
print(nome)
```

Veja que dentro da fun√ß√£o ``input()``, colocamos a mensagem que ir√° aparecer na tela, o que o usu√°rio digitar ser√° capturado e guardado dentro da vari√°vel "nome". Na linha em seguida, imprimos o que o usu√°rio digitou

Mas como podemos combinar a frase "hello world" com o nome digitado? Como vamos imprimir vari√°veis e textos juntos? Como podemos imprimir: "Ol√° Fulnao! Bem-vindo ao estudo de programa√ß√£o com Python. Parab√©ns pelo seu primeiro hello world"?  

Existe uma variedade de formas de imprimir texto e vari√°vel em Python. Vejamos algumas:  
Podemos usar formatadores de caracteres (igual em C);
Podemos usar a fun√ß√£o format();  
Podemos criar uma string formatada.  
Vejamos cada uma delas:

```python
# Modo 1 - usando formatadores de caracteres (igual na linguagem C) para imprimir vari√°vel e texto
print("Ol√° %s, bem vindo ao estudo de programa√ß√£o com Python. Parab√©ns pelo seu primeiro hello world" % (nome))
```

```python
# Modo 2 - usando a fun√ß√£o format() para imprimir vari√°vel e texto
print("Ol√° {}, bem vindo ao estudo de programa√ß√£o com Pyhton. Parab√©ns pelo seu primeiro hello world".format(nome))
```

```python
# Modo 3 - usando strings formatadas
print(f"Ol√° {nome}, bem vindo ao estudo de programa√ß√£o com Python. Parab√©ns pelo seu primeiro hello world")
```

Os tr√™s modos usados obtiveram o mesmo resultado e podemos adotar o que achar mais conveniente. Entretanto, existe um c√≥digo para sintaxe do "pythonic code". Pois bem, a PEP 498 (https://www.python.org/dev/peps/pep-0498/) fala sobre a cria√ß√£o de strings com interpola√ß√£o (mistura de vari√°veis com texto). Nessa PEP, o autor destaca o uso do **"modo 3"** como a melhor op√ß√£o chamando-a de **"f-strings"**. Portanto, em nossos c√≥digos, iremos adotar essa sintaxe para a cria√ß√£o de strings com interpola√ß√£o.

As strings formatadas com "f-strings" s√≥ podem ser compiladas com o interpretador Python da vers√£o 3.6 em diante. Se tentar usar essa sintaxe em um interpretador em vers√µes anteriores, ser√° dado erro de sintaxe.

Com rela√ß√£o √†s opera√ß√µes matem√°ticas, seja na programa√ß√£o, seja na resolu√ß√£o anal√≠tica, o mais importante √© lembrar a *ordem de preced√™ncias das opera√ß√µes.*

1. Primeiro resolvem-se os par√™nteses, do mais interno para o mais externo.
2. Exponencia√ß√£o.

√â fundamental conhecer a ordem de proced√™ncia das opera√ß√µes para n√£o criar c√°lculos errados durante a implementa√ß√£o de uma solu√ß√£o.

```python
# Qual o resultado armazendo na vari√°vel operacao_1: 25 ou 17?
operacao_1 = 2 + 3 * 5

# Qual o resultado armazendo na vari√°vel operacao_2: 25 ou 17?
operacao_2 = (2 + 3) * 5

# Qual o resultado armazendo na vari√°vel operacao_3: 4 ou 1?
operacao_3 = 4 / 2 ** 2

# Qual o resultado armazendo na vari√°vel operacao_4: 1 ou 5?
operacao_4 = 13 % 3 + 4
```

```python
print(f"Resultado em operacao_1 = {operacao_1}")
print(f"Resultado em operacao_2 = {operacao_2}")
print(f"Resultado em operacao_3 = {operacao_3}")
print(f"Resultado em operacao_4 = {operacao_4}")
```

Testando mais ainda.  
Uma equa√ß√£o do segundo grau possui a f√≥rmula:  
y = a*x**2 + b*x + c, onde a, b, c s√£o constantes.  
O valor de y (resultado) depende do valor de x, ou seja, x √© a vari√°vel independente e y a dependente.

Considerando os valores a = 2, b = 0.5 e c = 1, vamos solicitar para o usu√°rio um valor de x e retornar o valor de y correspondente ao x que ele digitou. Vejamoa:

```python
a = 2
b = 0.5
c = 1
x = input("Digite o valor de x: ")

y = a * x ** 2 + b * + c

print(f"O resultado de y para x = {x} √© {y}.")
```

Entendendo o erro.  
Vamos ler a mesagem de erro. Primeiro o erro foi do tipo "TypeError", isso quer dizer que alguma vari√°vel n√£o est√° com o tipo adequado para a situa√ß√£o. A mensagem nos diz que n√£o √© permitida a opera√ß√£o de potencia√ß√£o (**) entre um tipo string e inteiro. Portanto, o erro √© porque estamos tentando fazer uma opera√ß√£o matem√°tica entre string e um tipo num√©rico. Ent√£o, dentre as vari√°veis deve ter uma com o tipo errado. Vamos usar a fun√ß√£o ``type()`` para verificar o tipo das vari√°veis usadas.

```python
print(type(a))
print(type(b))
print(type(c))
print(type(x))
```

Olhando o resultado da entrada 18 (In [18]), vemos que o tipo da vari√°vel x √© string (str), isso acontece porque ao usar a fun√ß√£o ``input()``, ela retorna uma string, independente do que o usu√°rio digitou, sempre ser√° **string**. Portanto, para corrigir o erro, precisamos **converter** o resultado da entrada em um tipo **num√©rico**. Como n√£o sabemos se o usu√°rio vai digitar um n√∫mero inteiro ou decimal, vamos fazer a convers√£o usando a fun√ß√£o ``float()``. Veja como fica o c√≥digo:

```python
a = 2
b = 0.5
c = 1
x = input("Digite o valor de x: ")

x = float(x) # aqui fazemos a convers√£o da string para o tipo num√©rico

y = a * x ** 2 + b * + c

print(f"O resultado de y para x = {x} √© {y}.")
```
